```python
import pandas as pd 
from pandas_datareader import data 
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt 
import datetime

import re
```

```python
path="/content/drive/MyDrive/data/"
DATA = pd.read_csv("/content/drive/MyDrive/data/matala2_cosmetics_2019-Nov.csv")
df= DATA.copy()
```

# **Q.1**
'event_time' column to a datetime object

```python
df['event_time'] = df['event_time'].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S UTC'))
df.head()
```

Add to the dataframe a new column 'duration_to_next_event' that will hold the time in seconds from an event to the next event in the same visit.

```python
df.sort_values(['event_time','user_id','user_session'])
df['Time_of_the_next_event']=df.groupby(['user_session','user_id'])['event_time'].shift(-1)
df['duration_to_next_event'] = (df['Time_of_the_next_event']-df['event_time']).dt.total_seconds()
df['duration_to_next_event']=df['duration_to_next_event'].fillna(0)
df=df.drop('Time_of_the_next_event', axis=1)

df[df['user_session']=='0000d72e-be5f-4a03-9f17-6ac81584f177']	
```

# **Q.2**
Adding a column 'funnel_number' column - a sequence of visits by the same user to the site that are not more than 5 days apart.

```python
df['diff_days'] = df.groupby('user_id')['event_time'].diff().dt.days
df['panel'] = (df['diff_days'] > 5) | (df['diff_days'].isnull())
df['funnel_number'] = df.groupby('user_id') ['panel'].cumsum()
df=df.drop(['diff_days','panel'], axis=1)
df[df['user_id']==420448483]
```

# **Q.3**
Adding a column 'index_in_funnel' that will contain the number of visits to that panel

```python
df['index_in_funnel'] = df.groupby(['user_id', 'funnel_number']).cumcount() + 1
df[df['user_id']==420448483]
```

clean the 'price' column using regular expressions

```python
pattern = r'\d+(\.\d+)?'
df['price'] = df['price'].replace(to_replace=r'[^\d\.]', value='', regex=True)
df['price'] = df['price'].str.extract(pat=pattern).astype(float)
df.head()
```

# **Q.6**
The types of events that exist, a chart describing the number of events of each type


```python
grouped = df.groupby(['user_session'])

aggregations = {
    'duration_to_next_event': 'sum',
    'event_type': lambda x: list(x),
    'user_id': 'first',
    'funnel_number': 'first',
    'index_in_funnel': 'first',
    'product_id': lambda x: list(x),
    'category_code': lambda x: len(x),
    'product_id': lambda x: list(x)
}


session_data = grouped.agg(aggregations).reset_index()
session_data = session_data.rename(columns={'category_code': 'num_events'})
session_data = session_data.rename(columns={'duration_to_next_event': 'session_duration'})

session_data['list_of_viewed'] = session_data.apply(lambda x: [i for i, j in zip(x['product_id'], x['event_type']) if j == 'view'], axis=1)
session_data['list_of_added_to_cart'] = session_data.apply(lambda x: [i for i, j in zip(x['product_id'], x['event_type']) if j == 'cart'], axis=1)
session_data['list_of_purchased'] = session_data.apply(lambda x: [i for i, j in zip(x['product_id'], x['event_type']) if j == 'purchase'], axis=1)

session_data = session_data.drop(['product_id', 'event_type'], axis=1)
session_data
```
